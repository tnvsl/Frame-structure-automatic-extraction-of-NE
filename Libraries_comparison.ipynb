```python
import stanza
ner_stanza = []
nlp = stanza.Pipeline(lang='ru', processors='tokenize,ner', tokenize_no_ssplit=True)
List_column = DF['Предложение'].tolist()
doc = nlp(List_column)
```

```python
perEnts = []
for ent in doc.ents:
    if ent.type == "PER":
        perEnts.append(ent.text)
```

```python
locEnts = []
for ent in doc.ents:
    if ent.type == "LOC":
        locEnts.append(ent.text)
```

```python
locEnts
```

```python
orgEnts = []
for ent in doc.ents:
    if ent.type == "ORG":
        orgEnts.append(ent.text)
```

```python
len(orgEnts)
```

```python
orgEnts
```

```python
tokens = []
for sent in doc.sentences:
    for token in sent.tokens:
        tokens.append(token.text)
df = pd.DataFrame(tokens)
```

```python
df = df.rename(columns={0: 'origin'})
```

```python
ners = []
for sent in doc.sentences:
    for token in sent.tokens:
        ners.append(token.ner)
df['entities'] = ners
```

```python
df
```

```python
print(*[f'entity: {ent.text}\ttype: {ent.type}' for sent in doc.sentences for ent in sent.ents], sep='\n')
```

```python
print(*[f'token: {token.text}\tner: {token.ner}' for sent in doc.sentences for token in sent.tokens], sep='\n')
```
