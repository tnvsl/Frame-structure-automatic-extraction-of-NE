```python
import pandas as pd
import re
```

```python
df = pd.read_excel('/content/ruscorpora_expert_annotation.xlsx')
```

```python
raw_sentences = ' '.join(df['Предложение'])
```

```python
raw_sentences
```

```python
text = re.sub(r'\xa0', ' ', raw_sentences).strip()
```

```python
text
```

```python
import nltk
nltk.download('punkt')
sents = nltk.sent_tokenize(text)
```

LOC

```python
patternLoc = r'((?:город[а-я]{0,3}|улиц[а-я]{0,3}|(?:расположенн[а-я]{2}|происходивш[а-я]{2})?\s[Вв])\s(?:«?[А-ЯЁ]+[а-яё]*(?:[\s-][А-ЯЁ]+[а-яё]*)?))'
matchesLoc = re.findall(patternLoc, text)
print(matchesLoc)
len(matchesLoc)
```

```python
patternPlaces = r'(?:«?[А-ЯЁ]+[а-яё]*(?:[\s-][А-ЯЁ]+[а-яё]*)?)(?:\sплощад[а-я]?)'
matchesPlaces = re.findall(patternPlaces, text)
print(matchesPlaces)
len(matchesPlaces)
```

```python
patternLoc4Exclusion = r'(?:(?:город[а-я]{0,3}|(?:расположенн[а-я]{2}|происходивш[а-я]{2})?\s[Вв])\s(«?[А-ЯЁ]+[а-яё]*(?:[\s-][А-ЯЁ]+[а-яё]*)?))'
matchesLoc4Exclusion = re.findall(patternLoc4Exclusion, text)
print(matchesLoc4Exclusion)
```

### TIME

```python
patternYear = r"(\d{4}\s?г[.\S]{0,3})"
matchesYear = re.findall(patternYear, text)
print(matchesYear)
```

```python
patternDates = r'(\b(?:[1-9][0-9]?)\s(?:январ\S{1,3}|феврал\S{1,3}|март\S{0,2}|апрел\S{1,3}|ма\S{1,2}|июн\S{1,3}|июл\S{1,3}|август\S{0,2}|сентябр\S{1,3}|октябр\S{1,3}|ноябр\S{1,3}|декабр\S{1,3}))'
matchesDates = re.findall(patternDates, text)
print(matchesDates)
```

```python
patternFullDates = r'(\b(?:[1-9][0-9]?)\s(?:январ\S{1,3}|феврал\S{1,3}|март\S{0,2}|апрел\S{1,3}|ма\S{1,2}|июн\S{1,3}|июл\S{1,3}|август\S{0,2}|сентябр\S{1,3}|октябр\S{1,3}|ноябр\S{1,3}|декабр\S{1,3})\s(?:\d{4}\s?г[.\S]{0,3}))'
matchesFullDates = re.findall(patternFullDates, text)
print(matchesFullDates)
```

```python
patternPeriod = r'(с\s\b[1-9][0-9]?\s?(?:январ\S{1,3}|феврал\S{1,3}|март\S{0,2}|апрел\S{1,3}|ма\S{1,2}|июн\S{1,3}|июл\S{1,3}|август\S{0,2}|сентябр\S{1,3}|октябр\S{1,3}|ноябр\S{1,3}|декабр\S{1,3})?\sпо\s\b[1-9][0-9]\s(?:январ\S{1,3}|феврал\S{1,3}|март\S{0,2}|апрел\S{1,3}|ма\S{1,2}|июн\S{1,3}|июл\S{1,3}|август\S{0,2}|сентябр\S{1,3}|октябр\S{1,3}|ноябр\S{1,3}|декабр\S{1,3}))'
matchesPeriod = re.findall(patternPeriod, text)
print(matchesPeriod)
```

```python
patternHours = r'\b([1-9][0-9]?:\s?[0-9][0-9])'
matchesHours = re.findall(patternHours, text)
print(matchesHours)
```

```python
resTime = []
resTime = {'Years': matchesYear, 'Date + Month': matchesDates, 'FullDates': patternFullDates, 'Period': matchesPeriod, 'Hours': matchesHours}
print(resTime)
```

PRODUCTS

```python
resProdCovid = []
patternProdCovid = r'(?:(вакцин[а-я]?).+(«\S+»))'
for sent in sents:
    matchesProdCovid = re.findall(patternProdCovid, sent)
    if matchesProdCovid:
      for i in matchesProdCovid:
        res = ' '.join(i)
        resProdCovid.append(res)
print(resProdCovid)
```

```python
resProdRegistration = []
patternProdRegistration = r'(?:зарегистрирован[а-я]?|регистраци[а-я]?).+(«\S+»)'
for sent in sents:
    matchesProdRegistration = re.findall(patternProdRegistration, sent)
    if matchesProdRegistration:
      resProdRegistration.append(matchesProdRegistration)
print(resProdRegistration)
```

```python
resProd = []
resProd = {'Vaccine': resProdCovid, 'Registered Products': resProdRegistration}
print(resProd)
```

NUMBERS

```python
patternDec = r"(?<!\d)\d+(?:[.,]\d+)?(?!\d)"
matchesDec = re.findall(patternDec, text)

print(matchesDec)
matchesDecWOYears = [i for i in matchesDec if not re.findall(r'\d{4}', i)]
print(matchesDecWOYears)
```

```python
patternPercent = r'\d+%'
matchesPercent = re.findall(patternPercent, text)
print(matchesPercent)
```

```python
patternMeasures = r'\d+\s*(?:кг|мг|л|мл|мм|см|км|кгс)'
matchesMeasures = re.findall(patternMeasures, text)
print(matchesMeasures)
```

```python
resNum = []
resNum = {'Numbers': matchesDecWOYears, 'Percent': matchesPercent, 'Measures': matchesMeasures}
print(resNum)
```

ИСКУССТВО

```python
resCinema = []
patternCinema = r'([Фф]ильм\S{0,2}|[Сс]ериал\S{0,2}|[Мм]юзикл\S{0,2}).*([«"].+["»])'
for sent in sents:
    matchesCinema = re.findall(patternCinema, sent)
    if matchesCinema:
      for i in matchesCinema:
        res = ' '.join(i)
        resCinema.append(res)
print(resCinema)
```

```python
matchesArtist = []
patternArtist = r'(актер\S{0,2}|актрис\S{1,2}|певец\S{0,2}|певиц\S{1,2}|режиссер\S{0,2}|артистк?[а-я]{0,2}|автор[а-я]{0,2}|роль).*([А-ЯЁ][а-яё]+\s[А-ЯЁ][а-яё]+)'
for sent in sents:
    findArtist = re.findall(patternArtist, sent)
    if findArtist:
      for i in findArtist:
        res = ' '.join(i)
        matchesArtist.append(res)
print(matchesArtist)
```

```python
resArt = {}
resArt = {'Cinema': resCinema, 'Artists': matchesArtist}
print(resArt)
```
